```python
%matplotlib inline
import sklearn
# Import all of the scikit learn stuff
from __future__ import print_function
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import Normalizer
from sklearn import metrics
from sklearn.cluster import KMeans, MiniBatchKMeans
import pandas as pd
import numpy
```

```python
from os import listdir
from os.path import isfile, join
from ttp import ttp
import json
import csv
import pandas as pd

twi_path = 'C:/Users/hectorcflores1/Desktop/New folder/crowdsourced_neighborhoods-master/python/twitter/'
twi_jsons = [ f for f in listdir(twi_path) if isfile(join(twi_path,f)) ]


cnt = 0
all_ids = []
rows = []
for f in twi_jsons:
    with open('C:/Users/hectorcflores1/Desktop/New folder/crowdsourced_neighborhoods-master/python/twitter/%s' %(f)) as json_file:
        tweets = json.load(json_file).values()
        for tweet in tweets:
            content = tweet['content']
            p = ttp.Parser()
            parsed_tweet = p.parse(content)
            hashtags = [tag.encode('utf-8') for tag in parsed_tweet.tags]
            if float(tweet['lat']) != 0 and float(tweet['lon']) != 0:
                if tweet['tweet_id'] not in all_ids:
                    rows.append([tweet['lat'], tweet['lon'], tweet['time'], tweet['content'].encode('utf-8'), str(hashtags)])
                    cnt += 1

# build pandas dframe
tweets = pd.DataFrame(rows, columns =['lat','lon', 'time','tweet', 'tags'])
tweets.head()                       
```

```python

vectorizer = CountVectorizer(min_df = 1, stop_words = 'english')
dtm = vectorizer.fit_transform(tagSentence)
pd.DataFrame(dtm.toarray(),index=tagSentence,columns=vectorizer.get_feature_names
()).head(10)
```

```python
# Get words that correspond to each column
vectorizer.get_feature_names()
```

```python
#Fit LSA. Use algorithm = “randomized” for large datasets
lsa = TruncatedSVD(2, algorithm = 'randomized')
dtm_lsa = lsa.fit_transform(dtm)
dtm_lsa = Normalizer(copy=False).fit_transform(dtm_lsa)
```

```python
pd.DataFrame(lsa.components_,index = ["component_1","component_2"],columns =
vectorizer.get_feature_names())
```

```python
pd.DataFrame(dtm_lsa, index = tagSentence, columns = ["component_1","component_2"])
```

```python
xs = [w[0] for w in dtm_lsa]
ys = [w[1] for w in dtm_lsa]
xs, ys
```

```python
# Plot scatter plot of points
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(xs,ys)
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.title('Plot of points against LSA principal components')
```

```python
from matplotlib.pyplot import *
plt.figure()
ax = plt.gca()
ax.quiver(0,0,xs,ys,angles='xy',scale_units='xy',scale=1, linewidth = .01)
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
xlabel('First principal component')
ylabel('Second principal component')
title('Plot of points against LSA principal components')
plt.draw()
plt.show()
```

```python
# Compute document similarity using LSA components
similarity = np.asarray(numpy.asmatrix(dtm_lsa) * numpy.asmatrix(dtm_lsa).T)
pd.DataFrame(similarity,index=tagSentence, columns=tagSentence)
```
