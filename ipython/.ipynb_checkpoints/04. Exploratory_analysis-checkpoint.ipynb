# Exploratory Data Analysis

As mentioned, the data we used to analyze the 195 NYC neighborhoods was mostly obtained from the NYC Open Data website. In the first staged of the project, given that we did not  have enough social media data, we started by looking into basic statistical analysis of the urban data we had. 

We did this in various steps. First, we loooked into histograms of each variable and scatterplots to detect skewness in the data, possible correlations and their distribution. 

Secondly, we used supervised learning techniques, in particular, k-nearest neighbors in order to uncover possible clusterings of the NYC neighborhoods. We did Principal Component Analysis to reduce the dimensionality of our data and be able to visualize the result of our k-nearest neighbors implementation. The drawback of this approach, however, is that we had to input a classification for eac neighborhood given that they had none. The classification was made in 3 classes (very poor, somewhat poor and very poor) according to the average percentage of poverty perneighborhood.

Third, we used unsupevised learning, in particular, k-means. We did, like in the k-nearest neighbors algorithm, discover interesting clustering patterns, but given that k-means algorithm requires the user to input a (known) number of k clusters, we could not use this approach effectively.

Lastly, we implemented hierarchical clustering analysis. Hierarchical clusterting does not expect the user to input a known k number of clusters. Instead, it "discovers" clusters by itself by initializing random centroids and merging data points into an increasinly expanding number of clusters. Because we could not figure out what was a valid approach to define a better number of k clusters, however, we ultimately did not follow through this unsupervised learning approach.

## Dimensionality Reduction: Principal Components of NYC Neighborhoods Data

```python
import sklearn
import sklearn.cross_validation
import sklearn.decomposition
import sklearn.metrics
```

We performed PCA analysis in order to reduce the dimensionality of our data. In particular, we reduced our data into two principal components, which are linear combinations of all the data that explains the most variance. We did PCA in order to visualize possible clusters and perform the k-nearest neighbor algorithm. In order to make sure our 2 PCAs were actually explaining a significant level of variance, we used the explained variance ratio measure.

In order to perform the k-nearest neighbor algorith, we inputed a class to each neighborhood depending on its level of poverty. This classification was based on the non-transformed percentiles of the average percentage poverty data.

```python
# percentiles of AVG_PCT_POVERTY in order to create a classification

arr = np.array(data['AVG_PCT_POVERTY'])
np.percentile(arr, [0, 25, 50, 75, 100])
```

We created a target column for the dataframe, initialized in 0's, and then classify neigborhoods as not poor, somewhat poor and very poor. We use this classificiation to ultimately color tthe neighborhoods and identify clusters with the k-nearest neighbor algorithm.

```python
#create new column target, all 0
data['target'] = 0
```

```python
# assigning target variable to neighbohoods by % avg % of poverty
for i in range(data.shape[0]):
    if data['AVG_PCT_POVERTY'][i] <= 14.0: 
        data['target'][i] = 0 # not poor
    elif 14.0 < data['AVG_PCT_POVERTY'][i] >= 23.0:
        data['target'][i] = 1 #somewhat poor
    else:
        data['target'][i] = 2# very poor
```

```python
data.head(3)
```

```python
data.columns
```

We create a targe dataframe and a features dataframe in order to perform the k-nearest neighbors algorithm and later score it using F1 measures, which is a balanced measure of precision and recall for this type of supervised learning algorithms.

```python
#features without AVG_PCT_POVERTY
#removed all FAR data, except AVG_FAR

data_target = data['target']
data_features = data.loc[:, [ u'SUM_TOTAL_POP', u'AVG_TOTAL_SCORE', u'AVG_LOTARE', u'AVG_FAR',\
                             u'AVG_ASSESSEDVAL_TOTAL', u'AVG_PERCENT_OCC', u'AVG_MEDIAN_HHINCOME',\
                             u'AVG_MEANHH_HHINCOME', u'AVG_MEDHOME_VAL', u'AVG_MED_GROSS_RENTAL',\
                             u'Rent_pct_inc', u'AVG_UNITSTOTAL_WGHTD', u'CONSTRUCTIONNOISE_WGHTD', u'SUBWAYSTOPS_WGHTD',\
                             u'SHOOTINGS_WGHTD', u'GRAFFITI_WGHTD', u'DIRTY_WGHTD', u'DERELICTVEHICLES_WGHTD']]
print data_features.shape
data_features.head(3)
```

Before performing the k-nearest neighbor algorithm, we center the data by substracting to each data points its corresponding mean and dividing it by its standard deviation.

```python
"""
# normalized data
max_bycol = data_features.max(axis=0)
min_bycol = data_features.min(axis=0)
maxmin_bycol = max_bycol - min_bycol
feautures_norm = data_features / maxmin_bycol
feautures_norm.head(3)
"""
```

```python
features_centered = data_features - np.mean(data_features, axis=0)
features_centered = features_centered / np.std(data_features)
features_centered.head(3)
```

By looking at the ratio of the explained variance, we observe that our two first PCAs explain 38% of the overall variance in the NYC neighborhoods.

```python
# using PCA

# number of dimensions
n = data_features.shape[1]
# with PCA
#normalized features

from sklearn.decomposition import PCA
estimator = PCA(n_components=n)
x_pca = estimator.fit(features_centered)
x_pca.components_
# the percentage of explained variance ratio by principal component
x_pca_pve=x_pca.explained_variance_ratio_

print 'The Percentage of Explained Variance by the 2 first Principal\
 Components is', str(round(x_pca_pve[0] + x_pca_pve[1], 2)*100)+'%'
```

The sum of the explained variance should be 1.

```python
sum(x_pca_pve)
```

We plot the two PCAs and color the neighborhoods by the poverty classification discussed above. We can see that there is indeed a suggestive classification pattern. Without doing any further analytical research, we can see that it might be possible to use, for example, Support Vector Machines, to create hyperplanes that divide linearly clusters of neighborhood.

```python
# fitting centered data with PCA
x_pca_inplace= estimator.fit_transform(features_centered)

plt.figure(figsize=(10, 8))
plt.scatter(x_pca_inplace[:,0], x_pca_inplace[:,1], c=data_target, cmap=plt.cm.prism,  s= 45, alpha =0.5)
plt.title('Principal Components of NYC Neighborhood Data', size='x-large')
plt.xlabel('Principal Component 2', size='x-large')
plt.ylabel('Principal Component 1', size='x-large')
```

## Supervised Learning: KNN Classification for NYC Neighbors Data

We split randomly the data in order to create train and test data sets and be able to evaluate the performance of our k-nearest neighbor algorithm.

```python
#splitting data into training and test data

from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_features, data_target, test_size = 0.33, random_state=42)

print x_train.shape
print x_train
print y_train.shape
print y_train
```

We fit our algorithm, and look into the predicted classification.

```python
# create and fit a nearest neighbor classifier
from sklearn.neighbors import KNeighborsClassifier
knnCLF = KNeighborsClassifier()
knnCLF = knnCLF.fit(x_train, y_train)
```

```python
y_pred = knnCLF.predict(x_test)
y_pred
```

We calculate the score for our algorithm by validting it against our test data set.

```python
print "Score of KNN classifier with 5 neighbors (standard scoring):", knnCLF.score(x_test, y_test)    
```

We test different numbers of k-nearest neighbors in order to see what might be the best choice. We look both to the mean and std. dev of these scores.

```python
from sklearn.neighbors import KNeighborsClassifier
import sklearn.cross_validation


def knnScores(x_test, y_test, init, end):
    """returns panda df with cross validation
        score by neighbors, with mean and std dev"""
    scores= []
    for nNeighbors in range(init, end+1):
        knnCLF=  KNeighborsClassifier(n_neighbors= nNeighbors)
        score = sklearn.cross_validation.cross_val_score(knnCLF, x_test, y_test, cv=10, scoring ='f1') # why many scores?
        score_mean = np.mean(score)
        score_std = np.std(score)
        scores.append([nNeighbors, score_mean, score_std])

    print pd.DataFrame(scores, columns= ('Neighbors', 'Mean', 'Std. Dev.'))

# test
knnScores(x_test, y_test, 1, 20)
```

We visualize the same data than above using boxplots and observe that after7 or 8 neighbors, the scores tend to drop or behave irregularly. It would thus be recomendable to stick with a low number of neighbors in the actual algorithm implementation.

```python
from sklearn.neighbors import KNeighborsClassifier
import sklearn.cross_validation


def knnScores(x_test, y_test, init, end):
    """returns plot of cross validation scores
    vs. number of meighbors in KNN classifier"""
    scores= []
    for nNeighbors in range(init, end+1):
        knnCLF=  KNeighborsClassifier(n_neighbors= nNeighbors)
        score = sklearn.cross_validation.cross_val_score(knnCLF, x_test, y_test, cv=10, scoring ='f1') # why many scores?
        scores.append(score)

    plt.figure(figsize=(10, 6))
    plt.title('F1 Cross Validation Scores vs. Number of Neighbors', size= 'x-large')
    plt.xlabel('# of Estimators', size='x-large')
    plt.ylabel('Cross Validation Score', size= 'x-large')
    sns.boxplot(scores)

# test
knnScores(x_test, y_test, 1, 15)
```

# Unsupervised Learning: K-Means

We perform unsuperised learning using a k-mean algorithm. The disadvantage of this approach is that it requires the user to input a known umber of clusters. We ignore this number.

```python
from sklearn.cluster import KMeans
clf = KMeans(init='k-means++', n_clusters=5, random_state=42)
clf.fit(x_train)
```

```python
print clf.labels_.shape
print clf.labels_
```

We fit and visualize this algorithm testing different number of clusters. We use, again, the PCA data in order to fit this algorithm

```python
# Predict clusters on testing data
y_pred = clf.predict(x_test)
```

```python
# using PCA

# number of dimensions
n = data_features.shape[1]
# with PCA
#normalized features

from sklearn.decomposition import PCA
estimator = PCA(n_components=n)
x_pca = estimator.fit_transform(features_centered)
```

Using 3 clusters seems to be a better selection than using 5 clusters, as can be see in the plots, where data points are colored by cluster classification, and the centroids are shown with an X. 

```python
# with 2 k means

from sklearn import cluster

k = 3
data = x_pca

kmeans = cluster.KMeans(init='k-means++', n_clusters=k, random_state=42)
kmeans.fit(data)

#
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
#

plt.figure(figsize=(8, 6))

for i in range(k):
# select only data observations with cluster label == i
    ds = data[np.where(labels==i)]
    # plot the data observations
    plt.plot(ds[:,0],ds[:,1],'o')
    plt.title('K-Means Clusters for NYC Neighborhood Data (k=2)', size= 'x-large')
    plt.xlabel('Principal Component 2', size='x-large')
    plt.ylabel('Principal Component 1', size='x-large')
    # plot the centroids
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
    # make the centroid x's bigger
    plt.setp(lines,ms=15.0)
    plt.setp(lines,mew=2.0)
plt.show()
```

```python
# with 2 k means

from sklearn import cluster

k = 5
data = x_pca

kmeans = cluster.KMeans(init='k-means++', n_clusters=k, random_state=42)
kmeans.fit(data)

#
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
#

plt.figure(figsize=(8, 6))

for i in range(k):
# select only data observations with cluster label == i
    ds = data[np.where(labels==i)]
    # plot the data observations
    plt.plot(ds[:,0],ds[:,1],'o')
    plt.title('K-Means Clusters for NYC Neighborhood Data (k=2)', size= 'x-large')
    plt.xlabel('Principal Component 2', size='x-large')
    plt.ylabel('Principal Component 1', size='x-large')
    # plot the centroids
    lines = plt.plot(centroids[i,0],centroids[i,1],'kx')
    # make the centroid x's bigger
    plt.setp(lines,ms=15.0)
    plt.setp(lines,mew=2.0)
plt.show()
```
