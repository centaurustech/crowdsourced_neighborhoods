```python
from os import listdir
from os.path import isfile, join
from ttp import ttp
import json
import csv
import pandas as pd

twi_path = 'C:/Users/hectorcflores1/Desktop/New folder/crowdsourced_neighborhoods-master/python/twitter/'
twi_jsons = [ f for f in listdir(twi_path) if isfile(join(twi_path,f)) ]
#csv_file = open('csv/tweet_csv.csv', 'wt')

#writer = csv.writer(csv_file)
#writer.writerow(['lat', 'lon', 'time', 'content', 'tags'])

cnt = 0
all_ids = []
rows = []
for f in twi_jsons:
    with open('C:/Users/hectorcflores1/Desktop/New folder/crowdsourced_neighborhoods-master/python/twitter/%s' %(f)) as json_file:
        tweets = json.load(json_file).values()
        for tweet in tweets:
            content = tweet['content']
            p = ttp.Parser()
            parsed_tweet = p.parse(content)
            hashtags = [tag.encode('utf-8') for tag in parsed_tweet.tags]
            if float(tweet['lat']) != 0 and float(tweet['lon']) != 0:
                if tweet['tweet_id'] not in all_ids:
                    rows.append([tweet['lat'], tweet['lon'], tweet['time'], tweet['content'].encode('utf-8'), str(hashtags)])
                    cnt += 1
            #row = pd.DataFrame({'lat': , })
            #df.append(row)

tweets = pd.DataFrame(rows, columns =['lat','lon', 'time','tweet', 'tags'])
tweets.head()                       
```

```python
# mutate time into datetime format
tweets['time'] = pd.Series([pd.to_datetime(date) for date in tweets['time']])
tweets['hour'] =[tweets['time'][i].hour for i in range(len(tweets))]
tweets.head()
```

```python
"""
convertint string representation of list into an actual python list

http://stackoverflow.com/questions/1894269/convert-string-representation-of-list-to-list-in-python
"""
import ast

tweets['tags'] = [ast.literal_eval(tweets['tags'][row]) for row in range(len(tweets['tags']))]
#weets['tags'][0][0]
```

```python
# a list of all hastags as strings
allTags = []
for r in range(len(tweets['tags'])):
    if len(tweets['tags'][r]) !=0:
        for tag in range(len(tweets['tags'][r])):
            allTags.append(tweets['tags'][r][tag])
```

```python
len(allTags)
```

```python
from collections import Counter

#with location

def top_hashtags(hashtags):
    count= Counter(hashtags)
    top = count.most_common()[:25]
    tag = []
    freq = []
    for i in range(len(top)):
        tag.append(top[i][0])
        freq.append(top[i][1])
    plt.plot(freq)
    plt.xticks(range(1,26), tag, rotation='vertical', size='x-large')
    plt.ylabel('Count', size='x-large')

top_hashtags(allTags)
```
